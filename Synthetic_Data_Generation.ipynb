```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
NUM_GENERATORS = 500
```

```python
data = {
    "Generator": [f"G{i+1}" for i in range(NUM_GENERATORS)],
    "P_min": np.random.randint(20, 50, NUM_GENERATORS),
    "P_max": np.random.randint(100, 200, NUM_GENERATORS),
    "Cost_a": np.random.uniform(0.001, 0.005, NUM_GENERATORS),
    "Cost_b": np.random.uniform(5, 10, NUM_GENERATORS),
    "Cost_c": np.random.uniform(100, 300, NUM_GENERATORS),
    "Emission_coeff": np.random.uniform(0.2, 0.8, NUM_GENERATORS)
}
```

```python
df = pd.DataFrame(data)
df.head(5)
```

```python
df.to_csv("synthetic_generators.csv", index=False)
print("Synthetic generator data saved to synthetic_generators.csv")
print(df)
```
